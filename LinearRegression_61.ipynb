```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_squared_error
import math
```

```python
sal_data = pd.read_csv("/content/Salary_Data - Salary_Data.csv")
sal_data
```

```python
x = sal_data.drop('Salary',axis = 1)
y = sal_data.Salary
```

```python
train_x, test_x, train_y, test_y = tts(x,y,test_size=0.3, random_state=42)
```

```python
train_x.head()
```

```python
train_x, test_x, train_y, test_y = tts(x,y,test_size=0.3, random_state=30)
```

```python
train_x.head()
```

```python
test_x.head()
```

```python
train_y.head()
```

```python
test_y.head()
```

```python
model = LR()
```

```python
model.fit(train_x,train_y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
train_Y_hat = model.predict(train_x)
test_Y_hat = model.predict(test_x)
```

```python
print('Train MSE' , math.sqrt(mean_squared_error(train_y,train_Y_hat)))
print('Test MSE' , math.sqrt(mean_squared_error(test_y,test_Y_hat)))
```
